```python
import pandas as pd

df = pd.read_csv('./Life.csv')
df
```

```python
geo_df = pd.read_csv('./위도경도.csv')
geo_df
```

```python
# Country와 NAME이 같은 경우에 df2의 컬럼 추가
merged_df = pd.merge(df, geo_df, left_on='Country', right_on='NAME', how='left')

merged_df
```

```python
merged_df.to_csv('위도경도추가.csv', index=False)
```
