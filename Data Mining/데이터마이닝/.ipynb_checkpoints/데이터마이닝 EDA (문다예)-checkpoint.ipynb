```python
import pandas as pd

df = pd.read_csv('./Life Expectancy Data.csv')
df
```

```python
# 결측치 여부 확인
is_null = df.isnull()

# 각 열의 결측치 개수 확인
null_count = df.isnull().sum()
print(null_count)
```

# 결측치 제거
Population, GDP 변수 제거 후 결측치 제거

```python
df.drop(['Population', 'GDP'], axis=1, inplace=True)
```

```python
df = df.dropna()
df
```

### 범주형 데이터 타입 변경

```python
df.Status = df.Status.replace('Developing', 0)
df.Status = df.Status.replace('Developed', 1)
```

```python
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()
encoder.fit(df['Country'])
label = encoder.transform(df['Country'])
label = label.reshape(-1,1)
label
```

```python
df.drop(labels='Country', axis=1)
df['Country'] = label
```

```python
# 정규화하기
def normalize( x ) :
    return ( x - min(x) ) / ( max(x) - min(x) )

for i in range( len(df.columns) -2 ) :
    col = df.columns[i]
    df[col] = normalize( df[col] )
    
df
```

```python
df.columns = [column.strip() for column in df.columns]
df.columns
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# 종속 변수와 독립 변수 지정
dependent_variable = 'Life expectancy'
independent_variables = df.columns.difference([dependent_variable])
```

```python
independent_variables
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(df['Life expectancy'], kde=True)
plt.title('Distribution of Life Expectancy')
plt.show()

# 기대수명 분포: 60~80세 가장 많이 분포
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x='Status', y='Life expectancy', data=df)
plt.title('Boxplot of Life Expectancy by Status')
plt.show()

# 개발도상국과 선진국의 기대 수명 차이 분포 (boxplot)
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Adult Mortality', y='Life expectancy', data=df, hue='Status')
plt.title('Adult Mortality vs. Life Expectancy')
plt.show()

# Adult Mortality와 Life Expectancy' 상관관계
# 선진국일수록 Adult Mortality 낮은 특성을 보임
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Alcohol', y='Life expectancy', data=df, hue='Status', size='HIV/AIDS', sizes=(50, 200))
plt.title('Alcohol Consumption vs. Life Expectancy')
plt.show()

# 알콜 소비량과 기대수명 관계분석
# 알콜 소비량은 선진국이 더 많지만 기대수명은 높게 분포
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x=' BMI ', y='Life expectancy ', data=df, hue='Status')
plt.title('BMI vs. Life Expectancy')
plt.show()

# BMI 지수와 기대수명 관계
# 선진국이 BMI 지수가 높지만 기대수명이 더 높음
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='HIV/AIDS', y='Life expectancy', data=df, hue='Status')
plt.title('HIV/AIDS vs. Life Expectancy')
plt.show()

# 에이즈와 기대수명 관계 예측
# 에이즈 환자일수록 기대수명이 낮음
```

```python
correlation_matrix = df.corr()
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix[['Life expectancy']], annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation with Life Expectancy')
plt.show()

# status, alcohol, prcentage expenditure, Hepatitis B, BMI, polio, Total expenditure, Diphtheria, 
# 등등 기대수명과 상관관계가 있는 변수들로 보입니다 (양의 관계)
```
