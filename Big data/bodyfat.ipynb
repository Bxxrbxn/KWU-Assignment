```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style('dark')

df = pd.read_csv('bodyfat.csv', encoding = 'cp949')

W_quality = df.loc[df['Gender']=='W']
M_quality = df.loc[df['Gender']=='M']

sns.distplot(W_quality['BodyFat'], color='yellow', label='Woman')
sns.distplot(M_quality['BodyFat'], color='red', label='Man')

plt.title('Gender of BodyFat')
plt.legend()
plt.show()
```
